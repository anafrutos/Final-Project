```python
#adding continent name
import pycountry_convert as pc
def country_to_continent(country_name):
    try:
        country_alpha2 = pc.country_name_to_country_alpha2(country_name)
        country_continent_code = pc.country_alpha2_to_continent_code(country_alpha2)
        country_continent_name = pc.convert_continent_code_to_continent_name(country_continent_code)
    except:
        country_continent_name = 'Not assigned'
    return country_continent_name

countries['Continent'] = countries['country'].apply(lambda x:country_to_continent(x))
countries[countries['Continent'] == 'Not assigned']
countries.head()
```

```python
pd.options.display.max_rows=4629
BiocapTotGHA = data_with_index[data_with_index.record == 'BiocapTotGHA']
BiocapTotGHA[BiocapTotGHA.Continent == 'Europe']
```

```python
EFConsTotGHA = data_with_index[data_with_index.record == 'EFConsTotGHA']
```

```python
BiocapTotGHA = BiocapTotGHA.drop(['crop_land', 'grazing_land', 'forest_land',
       'fishing_ground', 'built_up_land', 'carbon'], axis=1)
```

```python
EFConsTotGHA = EFConsTotGHA.drop(['crop_land', 'grazing_land', 'forest_land',
       'fishing_ground', 'built_up_land', 'carbon'], axis=1)
```

```python
ECO = EFConsTotGHA.merge(BiocapTotGHA, how='inner', left_on=['country','year', 'Continent'], right_on = ['country','year', 'Continent'])
ECO.head()
```

```python
ECO = ECO.rename(columns= {'total_x':'TOT_EF','total_y':'TOT_Biocapacity'})
ECO
```

```python
ECO['Biocapacity_Res_Def'] = ECO.TOT_Biocapacity - ECO.TOT_EF
ECO
```

```python
Canada = ECO[ECO.country == 'Canada']
Canada.head()
```
